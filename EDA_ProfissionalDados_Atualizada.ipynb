## Qual o perfil geral do profissional da área de dados no Brasil?

#### Informações sobre gênero,idade , salário , estado , área de formação,Cargo Atual , Situação Atual de Trabalho:

```python
a = dados["('P1_b ', 'Genero')"].value_counts()
b = dados["('P1_b ', 'Genero')"].value_counts().plot(kind = 'bar' ,edgecolor = 'red',title = 'Gênero')
print(a,b)
```

##### Observa-se que a maioria dos profissionais da área de dados são do sexo masculino.

 ##### ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

```python
c = dados["('P1_a_a ', 'Faixa idade')"].value_counts()
d = dados["('P1_a_a ', 'Faixa idade')"].value_counts().plot(kind = 'barh' ,edgecolor = 'red',title = 'Idade')
print(c,d)
```

##### A Maioria dos Profissionais estão entre na faixa de 23 a 38 anos de idade, uma massa de profissionais jovens.

##### --------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

```python
e = dados["('P2_h ', 'Faixa salarial')"].value_counts()
f = dados["('P2_h ', 'Faixa salarial')"].value_counts().plot(kind = 'barh',edgecolor = 'red', title = 'Faixa Salarial')
print(e,f)
```

##### A maioria dos profissionais estão com a faixa salarial entre RS 8.000 a RS 12.000 reais por mês.

#####  -------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

```python
# Normalizando a faixa salarial
salamap = {    'Menos de R$ 1.000/mês': 1000,
      'de R$ 1.001/mês a R$ 2.000/mês': 1500, 
       'de R$ 2.001/mês a R$ 3000/mês': 2500,
      'de R$ 3.001/mês a R$ 4.000/mês': 3500,
      'de R$ 4.001/mês a R$ 6.000/mês': 5000, 
      'de R$ 6.001/mês a R$ 8.000/mês': 7000,
     'de R$ 8.001/mês a R$ 12.000/mês': 10000,
    'de R$ 12.001/mês a R$ 16.000/mês': 14000,
    'de R$ 16.001/mês a R$ 20.000/mês': 18000,
    'de R$ 20.001/mês a R$ 25.000/mês': 22500, 
    'de R$ 25.001/mês a R$ 30.000/mês': 27500,
    'de R$ 30.001/mês a R$ 40.000/mês': 35000,
    'Acima de R$ 40.001/mês':40000}
```

```python
# Criando a nova variável salário
dados['salario'] = dados["('P2_h ', 'Faixa salarial')"].map(salamap)
```

```python
# Agrupando os dados sobre o gênero,idade e salario
g = dados.groupby(["('P1_b ', 'Genero')","('P1_a_a ', 'Faixa idade')"]).mean()['salario'].unstack().T 
h =dados.groupby(["('P1_b ', 'Genero')","('P1_a_a ', 'Faixa idade')"]).mean()['salario'].unstack().T.plot(kind ='bar',edgecolor ='red',title ='Faixa de Idade , Gênero e Salário')
print(g.round(2),h)

```

###### Os mairores salários são pagos aos profissionais do sexo masculino com idade entre 45 a 49 anos .

###### ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

```python
i = dados["('P1_e ', 'Estado onde mora')"].value_counts()
j = dados["('P1_e ', 'Estado onde mora')"].value_counts().plot(kind = 'barh',edgecolor = 'red', title = 'Estado')
print(i,j)
```

###### A maior parte dos profissionais estão localizados no estado de São Paulo e nem todos os estados brasileiros possuem profissionais desse ramo.

##### --------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

```python
k = dados["('P1_i ', 'Área de Formação')"].value_counts()
l = dados["('P1_i ', 'Área de Formação')"].value_counts().plot(kind = 'barh',edgecolor = 'red', title = 'Área de Formação')
print(k,l)
```

#### Percebe-se que a grande maioria dos profissionais são formados nas áres de Computação,Engenharia de software, Sistemas de Informação e TI.

##### --------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

```python
l = dados["('P2_f ', 'Cargo Atual')"].value_counts()
m = dados["('P2_f ', 'Cargo Atual')"].value_counts().plot(kind ='barh',edgecolor ='red',title ='Cargo Atual')
print(l,m)
```

##### A maior partes dos profissionais estão trabalhando como cientista de dados.

#### ---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

```python
n = dados["('P2_a ', 'Qual sua situação atual de trabalho?')"].value_counts()
o = dados["('P2_a ', 'Qual sua situação atual de trabalho?')"].value_counts().plot(kind ='barh',edgecolor ='red',title ='Situação Atual de Trabalho')
print(n,o)
```
